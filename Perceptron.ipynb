```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def predict(X, weights):
    z = np.dot(X, weights)
    predictions = sigmoid(z)
    return predictions

def loss(X, Y, weights):
    # binary cross-entropy/log likelihood
    Y_ = predict(X, weights)
    cost = np.mean(-Y * np.log(Y_) - (1 - Y) * np.log(1 - Y_))
    return cost

def update_weights(X,Y,weights,learning_rate):
    Y_ = predict(X,weights)
    dw = np.dot(X.T,Y_ - Y)
    m = X.shape[0]
    weights = weights - learning_rate * dw / float(m)
    return weights

def train(X, Y, learning_rate, max_epochs):
    ones = np.ones((X.shape[0], 1))
    X = np.hstack((ones, X))
    weights = np.zeros(X.shape[1])
    
    for epoch in range(max_epochs):
        weights = update_weights(X, Y, weights, learning_rate)
        
        if epoch % 10 == 0:
            current_loss = loss(X, Y, weights)
            print("Epoch Number %d, Loss %.4f" % (epoch, current_loss))
        
    return weights
weigths=

    
```
